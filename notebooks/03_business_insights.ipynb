# Business Insights Analysis

This notebook extracts business insights from retail sales data.

# Imports

```python
import pandas as pd

df = pd.read_csv("../data/processed/clean_sales.csv")
```

# Top produtos

```python
top_products = (
    df.groupby("product_name")["sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```

# Clientes com maior faturamento

```python
top_customers = (
    df.groupby("customer_name")["sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_customers
```

# Categoria mais lucrativa

```python
category_sales = (
    df.groupby("category")["sales"]
    .sum()
    .sort_values(ascending=False)
)

category_sales
```

## Key Insights

- Technology products generated the highest sales volume.
- The West region showed strong sales performance.
- Sales peaks occurred during year-end periods.
